Plotting a line chart from a Pandas object

```python
address = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')
address
```

```python
cars = address

mpg = cars['mpg']
mpg.plot();
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot();
```

```python
color_theme = ['darkgray', 'lightsalmon', 'powderblue']
df.plot(color=color_theme)
```

```python
mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.model, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('Car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```

```python
mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.model, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('Car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')

ax.set_ylim([0,45])

ax.annotate('Toyota Corolla', xy= (19, 33.9), xytext=(21,35),
            arrowprops=dict (facecolor='black', shrink=0.05))
```

Creating a pie chart

```python
z = [1,2,3,4,0.5]
plt.pie(z)
plt.show()
```

Saving a plot

```python
plt.pie(z)
plt.savefig('pie_chart.png')
plt.show()
```

```python
color_theme = ['#A949A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
plt.pie(z, colors=color_theme)
plt.show()
```

```python
veh_type = ['bycicle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels=veh_type)
plt.show()
```

```python
# Adding a legend to your plot

plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```
